```python
import random
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from fredapi import Fred

# Set Reproducible Seed
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
set_seed(42)
# Config
FRED_API_KEY = "f67b94cab05139f9fdc47ddf41c702db"
INPUT_WINDOW = 60
OUTPUT_WINDOW = 7
LATENT_DIM = 128
N_CLUSTERS = 5
BATCH_SIZE = 64
AE_EPOCHS = 30
LSTM_EPOCHS = 30
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# FRED Data Fetching
fred = Fred(api_key=FRED_API_KEY)
indicators = {
    'CPIAUCSL': 'cpi',
    'DCOILWTICO': 'oil_price',
    'FEDFUNDS': 'fed_funds_rate',
    'UNRATE': 'unemployment',
    'INDPRO': 'industrial_production',
    'UMCSENT': 'consumer_sentiment',
    'PCE': 'personal_consumption',
    'GS10': 'ten_year_treasury_rate',
    'CE16OV': 'employment_level',
    'M2SL': 'm2_money_supply'
}
fred_data = pd.DataFrame()
for series_id, col_name in indicators.items():
    series = fred.get_series(series_id, observation_start='2010-01-01')
    tmp = series.reset_index()
    tmp.columns = ['date', col_name]
    fred_data = tmp if fred_data.empty else pd.merge(fred_data, tmp, on='date', how='outer')

fred_data = fred_data.sort_values('date').fillna(method='ffill').fillna(method='bfill')
fred_data.set_index("date", inplace=True)

# Normalize
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(fred_data)

# Create Sequences
def create_sequences(data, window):
    return np.array([data[i:i+window] for i in range(len(data) - window)])
sequences = create_sequences(scaled_data, INPUT_WINDOW)
sequences_tensor = torch.tensor(sequences, dtype=torch.float32).to(DEVICE)
# Autoencoder
class Autoencoder(nn.Module):
    def __init__(self, input_dim, window_size, latent_dim):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Flatten(),
            nn.Linear(input_dim * window_size, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, input_dim * window_size)
        )

    def forward(self, x):
        z = self.encoder(x)
        recon = self.decoder(z).view(x.size())
        return recon, z

autoencoder = Autoencoder(scaled_data.shape[1], INPUT_WINDOW, LATENT_DIM).to(DEVICE)
ae_optimizer = torch.optim.Adam(autoencoder.parameters(), lr=0.001)
ae_loss_fn = nn.MSELoss()

ae_loader = DataLoader(TensorDataset(sequences_tensor), batch_size=BATCH_SIZE, shuffle=True)
for epoch in range(AE_EPOCHS):
    total_loss = 0
    for (xb,) in ae_loader:
        ae_optimizer.zero_grad()
        recon, _ = autoencoder(xb)
        loss = ae_loss_fn(recon, xb)
        loss.backward()
        ae_optimizer.step()
        total_loss += loss.item()
    print(f"[AE] Epoch {epoch+1}, Loss: {total_loss:.4f}")

# KMeans Clustering
autoencoder.eval()
with torch.no_grad():
    _, embeddings = autoencoder(sequences_tensor)
embeddings_np = embeddings.cpu().numpy()
cluster_labels = KMeans(n_clusters=N_CLUSTERS, random_state=42).fit_predict(embeddings_np)

# Forecasting Data
def create_forecast_data(data, input_window, output_window):
    X, y = [], []
    for i in range(len(data) - input_window - output_window):
        X.append(data[i:i+input_window])
        y.append(data[i+input_window:i+input_window+output_window, 0])
    return np.array(X), np.array(y)

X_seq, y_seq = create_forecast_data(scaled_data, INPUT_WINDOW, OUTPUT_WINDOW)
enc = OneHotEncoder(sparse_output=False)
clusters_encoded = enc.fit_transform(cluster_labels[:len(X_seq)].reshape(-1, 1))
clusters_expanded = np.repeat(clusters_encoded[:, np.newaxis, :], INPUT_WINDOW, axis=1)
X_aug = np.concatenate([X_seq, clusters_expanded], axis=2)
X_tensor = torch.tensor(X_aug, dtype=torch.float32).to(DEVICE)
y_tensor = torch.tensor(y_seq, dtype=torch.float32).to(DEVICE)
split = int(0.8 * len(X_tensor))
X_train, X_test = X_tensor[:split], X_tensor[split:]
y_train, y_test = y_tensor[:split], y_tensor[split:]
train_loader = DataLoader(TensorDataset(X_train, y_train), batch_size=BATCH_SIZE, shuffle=True)

class CNNLSTMForecast_Improved(nn.Module):
    def __init__(self, input_size, cnn_channels=64, lstm_hidden=128, output_size=OUTPUT_WINDOW, dropout_rate=0.2):
        super(CNNLSTMForecast_Improved, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=input_size, out_channels=cnn_channels, kernel_size=3, padding=1)
        self.relu = nn.ReLU()

        self.lstm1 = nn.LSTM(input_size=cnn_channels, hidden_size=lstm_hidden, batch_first=True)
        self.lstm2 = nn.LSTM(input_size=lstm_hidden, hidden_size=lstm_hidden, batch_first=True)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(lstm_hidden, output_size)

    def forward(self, x):
        x = x.permute(0, 2, 1)
        x = self.relu(self.conv1(x))
        x = x.permute(0, 2, 1)
        out, _ = self.lstm1(x)
        out, _ = self.lstm2(out)
        out = self.dropout(out[:, -1, :])
        return self.fc(out)
model = CNNLSTMForecast_Improved(input_size=X_train.shape[2]).to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
loss_fn = nn.MSELoss()

# Train
for epoch in range(LSTM_EPOCHS):
    model.train()
    total_loss = 0
    for xb, yb in train_loader:
        optimizer.zero_grad()
        pred = model(xb)
        loss = loss_fn(pred, yb)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"[CNN+LSTM] Epoch {epoch+1}, Loss: {total_loss:.4f}")

# Evaluate & Compare
model.eval()
with torch.no_grad():
    preds = model(X_test).cpu().numpy()
    targets = y_test.cpu().numpy()

# Denormalize
cpi_min = scaler.data_min_[0]
cpi_max = scaler.data_max_[0]
preds_denorm = preds[:, 0] * (cpi_max - cpi_min) + cpi_min
targets_denorm = targets[:, 0] * (cpi_max - cpi_min) + cpi_min

# Accuracy
mae = mean_absolute_error(targets_denorm, preds_denorm)
accuracy = 1 - mae / np.mean(targets_denorm)
print(f"\n Mean Absolute Error: {mae:.4f}")
print(f" Estimated Accuracy: {accuracy * 100:.2f}%")

#Table Comparison
num_examples = 10
comparison_df = pd.DataFrame({
    "Day": list(range(1, num_examples + 1)),
    "Actual CPI (Day 1)": targets_denorm[:num_examples],
    "Predicted CPI (Day 1)": preds_denorm[:num_examples],
    "Error": np.abs(targets_denorm[:num_examples] - preds_denorm[:num_examples])
})

comparison_df["% Error"] = (comparison_df["Error"] / comparison_df["Actual CPI (Day 1)"]) * 100
print("\n Comparison Table (Denormalized):")
display(comparison_df.round(2))

# Bar Chart

labels = [f'Day {i}' for i in range(1, num_examples + 1)]
x = np.arange(num_examples)
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar(x - width/2, comparison_df["Actual CPI (Day 1)"], width, label='Actual')
bars2 = ax.bar(x + width/2, comparison_df["Predicted CPI (Day 1)"], width, label='Predicted')
for i in range(num_examples):
    err = comparison_df["% Error"].iloc[i]
    ax.text(x[i], max(bars1[i].get_height(), bars2[i].get_height()) + 0.5, f"{err:.1f}%", ha='center', fontsize=9)

ax.set_xlabel("Test Sample (Day Index)")
ax.set_ylabel("CPI Value")
ax.set_title(" Actual vs. Predicted CPI (Day 1 of Forecast Window)")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.tight_layout()
plt.show()

# Plot results
plt.figure(figsize=(15, 8))
for i in range(min(5, len(preds))):
    plt.subplot(5, 1, i + 1)
    plt.plot(targets[i], label="Actual", marker='o')
    plt.plot(preds[i], label="Predicted", marker='x')
    plt.title(f"Forecast Sequence {i+1}")
    plt.xlabel("Day Ahead")
    plt.ylabel("Normalized CPI")
    plt.legend()
plt.tight_layout()
plt.show()
```
